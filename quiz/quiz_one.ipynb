## Calculate the circumference of a circle

```python
def circumference_calculate():
    try:
        radius = input("Please enter an integer:\n")
        print('\n')
        print('circumference is %f' % (math.pi * 2 * float(radius)))
    except ValueError or NameError:
        print('\n')
        print('Please enter an INTEGER!')
```

```python
circumference_calculate()
```

```python
circumference_calculate()
```

## Financial Compound Model

```python
def compound_model():
    import math
    try:
        principal = float(input("Please enter your principal:\n"))
        interest = float(input("Please enter your interest rate:\n"))
        years = float(input("Please enter number of years:\n"))
        print ("\n")
        print('Your maturity value is %f' % (principal*((1+interest/12)**(years*12))))
    except ValueError or NameError:
        print ("\n")
        print('Please make sure to enter correct values!')
```

```python
compound_model()
```

```python
compound_model()
```

## Grade Calculator

```python
def grade_calculator():
    score = int(input("Please give your grade:\n"))
    
    if score >= 95:
        grade = 'A+'
    elif score >= 90:
        grade = 'A'
    elif score >= 87:
        grade = 'A-'
    elif score >= 85:
        grade = 'B+'
    elif score >= 80:
        grade = 'B'
    else:
        grade = 'B-'
    print('Your grade is %s'% grade)
```

```python
grade_calculator()
```
